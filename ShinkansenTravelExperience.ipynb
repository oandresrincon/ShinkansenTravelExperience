```python
categorical_columns = ['Seat_Class', 'Gender', 'Customer_Type', 'Type_Travel', 'Travel_Class']
```

```python
def one_hot_encode(col_list, dataframe):
    for column in col_list:
        dataframe[column].fillna('sin_valor', inplace=True)
        # Get one hot encoding of column
        one_hot = pd.get_dummies(dataframe[column], prefix=column)
        # Drop column as it is now encoded
        dataframe = dataframe.drop(column,axis = 1)
        # Join the encoded df
        dataframe = dataframe.join(one_hot)
    return dataframe
data = one_hot_encode(categorical_columns,data)
```

### **Filling null values in the dataset with means**

```python
def fill_mean(data,col_names):
    for col_name in col_names:
        if col_name in ['Age','Travel_Distance' ,'Departure_Delay_in_Mins','Arrival_Delay_in_Mins']:
            data[col_name].fillna(int(data[col_name].mean()), inplace=True)#data[col_name].mean()
        else:
            data[col_name].fillna(int(data[col_name].mode()), inplace=True)#data[col_name].mean()
    return data
cols_list = data.select_dtypes(include=np.number).columns.tolist()
cols_list.remove('Overall_Experience')
data=fill_mean(data,cols_list)
```

### Function to clean the data

```python
def clean_data(data):
    cols_list_1 = data.columns.tolist()
    if 'Overall_Experience' in cols_list_1:
        Overall_Experience=data['Overall_Experience']    
    categorical_columns = ['Seat_Class', 'Gender', 'Customer_Type', 'Type_Travel', 'Travel_Class']
    data = one_hot_encode(categorical_columns,data)
    data=encode_cols_list_similar(cols_list_similar,data)
    data['Platform_Location'].replace(['Very Inconvenient', 'Inconvenient', 'Needs Improvement', 'Manageable', 'Convenient', 'Very Convenient'],
                             [0, 1, 2, 3, 4, 5], inplace=True)
    data['SurveyPos'] = data.apply(lambda row: ((#row.Platform_Location +
                                             row.Seat_Comfort +
                                             #row.Arrival_Time_Convenient +
                                             #row.Catering +
                                             #row.Onboard_Wifi_Service +
                                             row.Onboard_Entertainment 
                                             #row.Online_Support +
                                             #row.Ease_of_Online_Booking 
                                             #row.Onboard_Service +
                                             #row.Legroom +
                                             #row.Baggage_Handling +
                                             #row.CheckIn_Service +
                                             #row.Cleanliness +
                                             #row.Online_Boarding
                                             )/2), axis=1)
    data['SurveyNeg']=data['SurveyPos']
    
    #DataSurveyPos[DataSurveyPos>=3]=1
    #DataSurveyPos[DataSurveyPos<3]=0
    #data['SurveyPos']=DataSurveyPos
    data['SurveyPos'].values[data['SurveyPos'].values <= 3.0] = 0
    data['SurveyPos'].values[data['SurveyPos'].values > 3.0] = 1
    data['SurveyNeg'].values[data['SurveyNeg'].values <= 2.0] = 0
    data['SurveyNeg'].values[data['SurveyNeg'].values > 2.0] = 1    
    cols_list = data.select_dtypes(include=np.number).columns.tolist()
    #cols_list.remove('Overall_Experience')
    data=norm_cols(cols_list,data)
    data=fill_mean(data,cols_list)
    if 'Overall_Experience' in cols_list_1:
        data['Overall_Experience'] = Overall_Experience    
    return data
```
